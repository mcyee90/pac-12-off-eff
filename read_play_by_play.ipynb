```python
schedule = pd.read_csv("schedule.csv")
play_list = []
for i, r in schedule.iterrows():
    team = r['Team']
    abbr = r['Abbreviation']
    opp = r['Opponent']
    url = "https://www.ncaa.com/" + r['link'] + "/play-by-play"
    print(url)
    executable_path = {'executable_path': 'chromedriver.exe'}
    browser = Browser('chrome', **executable_path, headless = True)
    browser.visit(url)
    game_html = browser.html
    game_soup = BeautifulSoup(game_html, "html.parser")
    game_table = game_soup.findAll("div", {"class": "play-by-play-period-football-table"})
#     print(len(game_table))
    for x in np.arange(len(game_table)):
        try:
            possession = game_table[x].find_all("th", {"class":"football-description"})
            plays = game_table[x].find_all("tbody")
            y = 0
            for pos in possession:
                if team + " at" in pos.text:
#                     print (y, pos.text)
                    for p in plays[y].find_all("tr"):
                        for d in p.find_all("td", {"class": "drive"}):
                            dd = d.text.replace('\n', '')
                        for pp in p.find_all("td", {"class":"football-description"}):
                            play = pp.text.replace('\n', '')
                        play_dict = {"team": team,
                            "abbr": abbr,
                            "opponent": opp,
                            "Date": r['Date'],
                            "HA": r['HA'],
                            "down_distance": dd,
                            "play": play,
                            "wk":r['week']}
                        play_list.append(play_dict)
                y=y+1
    #         print(play_list)
        except:
            pass

    print(opp + " vs. " + team + " scraped!")
```
