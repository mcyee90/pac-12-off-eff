```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
import requests
import numpy as np
import time
import json
pd.options.display.max_rows = 999
```

```python
weeks = ["01","02","03","04","05","06","07","08","09","10","11","12","13","14","P"]
schedule = []
for n in weeks:
    url = "https://www.ncaa.com/scoreboard/football/fbs/2018/" + n + "/pac-12"
    executable_path = {'executable_path': 'chromedriver.exe'}
    browser = Browser('chrome', **executable_path, headless = True)
    browser.visit(url)
    schedule_html = browser.html
    schedule_soup = BeautifulSoup(schedule_html, "html.parser")
    schedule_table = schedule_soup.findAll("div", {"class": "gamePod-type-game"})
#     print(schedule_table)

    for x in np.arange(len(schedule_table)):
        for a in schedule_table[x].find_all("a", href=True):
            link = a['href']
        teams = schedule_table[x].find_all("span", {"class":"gamePod-game-team-name"})
        a_team = teams[0].text
        h_team = teams[2].text
        scores = schedule_table[x].find_all("span", {"class":"gamePod-game-team-score"})
        a_score = scores[0].text
        h_score = scores[1].text
        h_dictionary = {"week":n,
              "link":link,
              "Team":h_team,
              "Opponent":a_team,
              "HA":"H",
              "Date": link.split("/")[5] + "/" + link.split("/")[6] + "/" + link.split("/")[4],
              "pf": h_score,
              "pa": a_score}
        a_dictionary = {"week":n,
              "link":link,
              "Team":a_team,
              "Opponent":h_team,
              "HA":"A",
              "Date": link.split("/")[5] + "/" + link.split("/")[6] + "/" + link.split("/")[4],
              "pf": a_score,
              "pa": h_score}

        schedule.append(h_dictionary)
        schedule.append(a_dictionary)
        print(a_team, a_score, "vs.", h_team, h_score)
    print("Week " + n + " complete!")
```

```python
schedule_df = pd.DataFrame(schedule)
schedule_df
```

```python
pac_12_teams = ["Arizona", "Arizona St.", "California", "Colorado", "Oregon", "Oregon St.", 
                "Southern California", "Stanford", "UCLA", "Utah", "Washington", "Washington St."]
abbreviations = ["ARI", "ASU", "CAL", "COL", "ORE", "ORS",
                "USC", "STA", "UCLA", "UTH", "WAS", "WST"]
```

```python
for i, r in schedule_df.iterrows():
    if r['Team'] not in pac_12_teams:
        schedule_df = schedule_df.drop(i)
```

```python
schedule_df['game_id'] = ""
for i, r in schedule_df.iterrows():
    schedule_df.at[i, 'game_id'] = r['week']+r['Team']
schedule_df
```

```python
teams_df = pd.DataFrame({"Team":pac_12_teams,
           "Abbreviation": abbreviations})
```

```python
teams_df
```

```python
schedule_df = schedule_df.merge(teams_df, left_on='Team', right_on='Team', how='left')
schedule_df = schedule_df.drop(128)
schedule_df = schedule_df.drop(129)
schedule_df = schedule_df.reset_index(drop=True)
```
